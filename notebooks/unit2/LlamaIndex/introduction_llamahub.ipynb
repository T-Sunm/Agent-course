```python
from llama_index.llms.ollama import Ollama
import asyncio
import nest_asyncio
nest_asyncio.apply()


# Khởi tạo Ollama LLM
llm = Ollama(
    model="deepseek-r1:1.5b",   # Đảm bảo model này đã được cài đặt trên server Ollama
    host="127.0.0.1",           # Địa chỉ host (mặc định của server Ollama)
    port=12345,                 # Port mà server Ollama đang lắng nghe
    request_timeout=360.0       # Thời gian timeout cho request
)

async def generate_text(prompt: str):
    # Gửi prompt đến Ollama và nhận phản hồi
    response = await llm.complete(prompt)
    print("Response:", response)

# Gọi hàm generate_text với prompt mẫu
await generate_text("Hello, how are you?")
```

```python
import asyncio
import nest_asyncio
import requests
from llama_index.llms.ollama import Ollama
from llama_index.core.agent.workflow import FunctionAgent

nest_asyncio.apply()

# Kiểm tra kết nối đến Ollama server
try:
    r = requests.get("http://127.0.0.1:12345")
    print("Kết nối đến Ollama server thành công, status code:", r.status_code)
except Exception as e:
    print("Không thể kết nối đến Ollama server:", e)

# Định nghĩa một công cụ tính toán đơn giản
def multiply(a: float, b: float) -> float:
    """Useful for multiplying two numbers."""
    return a * b


# Tạo workflow agent với công cụ calculator
agent = FunctionAgent(
    tools=[multiply],
    llm=Ollama(
        model="deepseek-r1:1.5b",  # Hãy chắc chắn model này có sẵn trên server
        host="http://127.0.0.1",
        request_timeout=360.0
    ),
    system_prompt="You are a helpful assistant that can multiply two numbers.",
)

async def main():
    # Chạy agent
    response = await agent.run("What is 1234 * 4567?")
    print("Kết quả của agent:", response)

# Chạy hàm main() trong Jupyter
await main()
```
